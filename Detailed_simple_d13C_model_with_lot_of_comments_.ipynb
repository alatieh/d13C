Finding out the 13C of the three main fossil fuels (coal,oil and gas) 

```python
# The average d13C value of every fossil fuel type
# unit is permil VPDB
# It is very rough estimation because: for example for Gas there are three main types (conventional gas, shale gas, and coal gas) beside the extraction and the geological layer that difference
# I still need to read thoroughly the literature that I get from Thomas Roeckmann about the d13C value 

methane_13C = -45.2   # This is the average of the natural gas/coal from study database https://essd.copernicus.org/articles/14/4365/2022/essd-14-4365-2022.pdf    page 13, table5 , I need to find the better value of the fossil CH4 only
coal_13C= -28   # I got this number from https://gml.noaa.gov/outreach/isotopes/c13tellsus.html but I need more exact number (.........working on)
oil_13C= -28  # I need to find out the real number but this just for my back of the envelope calculation # the number of -28 also found in https://gml.noaa.gov/ccgg/isotopes/c13tellsus.html
```

We have all the ingredients of reservoirs and fluxes. Let us run the system 

```python
# making a loop for the atmosphere
#  d13C X new atmosphere = (d13C X old atmosphere)   +   (d13C X fluxes)
# rearrange it
#  d13C  = [(d13C X old atmosphere)   +   (d13C X flux)]    /  new\atmosphere

# we have (in and out) flux for biosphere and ocean
# we have in flux from fossil fuel
# so in total 5 fluxes 
# the loop is 52 steps

atmosphere_13C =  -7   # a begin value of atmosphere
biosphere_13C = -25    #  a begin value
ocean_13C = 1.8        #  a begin value


atmosphere_13C_list=[]
biosphere_13C_list=[]
ocean_13C_list=[]

# The fossil fuel flux is 10 000 MtC times  -28 permill  = -280 000

# The first loop is to fill the list of atmosphere_13C_list
# Then the second part of the loop to fill the biosphere and ocean lists
for i in range(52):
  atmosphere_13C = (atmosphere*atmosphere_13C +\
  from_biosphere_to_Atmosphere * biosphere_13C -\
  from_Atmosphere_to_biosphere * atmosphere_13C  +\
  from_ocean_to_Atmosphere*ocean_13C -\
  from_Atmosphere_to_ocean* atmosphere_13C-\
  280000)/(atmosphere +\
  from_biosphere_to_Atmosphere  -\
  from_Atmosphere_to_biosphere   +\
  from_ocean_to_Atmosphere -\
  from_Atmosphere_to_ocean)
  #print(atmosphere_13C)
  atmosphere_13C_list.append(atmosphere_13C)

plt.plot(time, atmosphere_13C_list)
plt.show()
```

```python
biosphere_13C_list=[]
biosphere_13C = -25    #  a begin value

for i in range(52): 
  biosphere_13C = (biosphere*biosphere_13C-\
  from_biosphere_to_Atmosphere*biosphere_13C+\
  from_Atmosphere_to_biosphere*atmosphere_13C)/(biosphere -from_biosphere_to_Atmosphere+from_Atmosphere_to_biosphere)
  biosphere_13C_list.append(biosphere_13C)
  #print(biosphere_13C)

plt.plot(time,biosphere_13C_list)
plt.show()
```

```python
ocean_13C_list=[]
ocean_13C = 1.8     #  a begin value

for i in range(52): 
  ocean_13C = (ocean*ocean_13C-\
  from_ocean_to_Atmosphere*ocean_13C+\
  from_Atmosphere_to_ocean*atmosphere_13C)/(ocean -from_ocean_to_Atmosphere+from_Atmosphere_to_ocean)
  ocean_13C_list.append(ocean_13C)
  #print(ocean_13C)

plt.plot(time,ocean_13C_list)
plt.show()
```

```python
plt.plot(time, atmosphere_13C_list, label='atmosphere')
plt.plot(time,biosphere_13C_list, label='biosphere')
plt.plot(time,ocean_13C_list, label='ocean')
plt.xlabel("year")
plt.ylabel("δ13C  ‰")
plt.title("Simple model of δ13C")
plt.legend()
plt.show()
```
